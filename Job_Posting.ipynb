# Analyse no IT

#### ici nous allons procder au prtraitement type tokenisazion lemnanisation et suppression des stop word pour ensuite pouvoir avoir une reprsentation vectorielle de chaque job offer

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
ps = PorterStemmer()
data_w_token = word_tokenize(all_text_no_IT) 
data_w_token = [w.lower() for w in data_w_token if len(w) > 3]
```

```python
stop_words = set(stopwords.words('english'))
data_Nstop_word = [w for w in data_w_token if  w  not in stop_words] #suppresion des stopwords
data_normalized = [ps.stem(w) for w in data_Nstop_word] #racination
data_normalized = [w for w in data_normalized if  w  not in stop_words] #peut tre que des mots n'taient pas dans les stop word avant d'tre lemnis
```

#### Creation des features qui permettrons l'initialisation de nos matrice bag-of-words

```python
features = []
for i in range(len(data_normalized)):
    if data_normalized[i] not in features:
        features.append(data_normalized[i])

```

#### Nettoyage de texte et mise en forme de matrice de mots pour pouvoir en extraires de matrice de type bag-of-words

```python
X = []
for i in range(len(data_no_IT)):
    if i > 0:
        string = word_tokenize(string)
        string = [w.lower() for w in string if len(w) > 3]
        string = [w for w in string if  w  not in stop_words]
        string = [ps.stem(w) for w in string]
        X.append(string)
        string = ''
    else:
        liste = []
        string = ''
    for j in range(len(data_no_IT[i])):
            mot = str(data_no_IT[i][j])
            mot = mot.replace('\r', '')
            mot = mot.replace('\n', '')
            mot = mot.replace('-', '') 
            string += mot
            
```

#### Appelle de diffrentes librairies, ici nous allons d'abord compter combien de mots chaque job offer possde par features  l'aide d'une collection puis transformer ce comptage en matrice qui compte le nombre d'occurence

```python
from sklearn.feature_extraction import DictVectorizer
from collections import Counter, OrderedDict



v = DictVectorizer()
v.fit([OrderedDict.fromkeys(features, 1)])
X = v.transform(Counter(f) for f in X)

```

#### l'ACP n'est pas possible pour une matrice  faible densit nous allons donc la rendre plus dense grce  random_projection

```python
from sklearn import random_projection
transformer = random_projection.SparseRandomProjection()
X_new = transformer.fit_transform(X)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X2D = pca.fit_transform(X_new.A)
print(pca.explained_variance_ratio_)
```

```python
pca.explained_variance_ratio_.sum()
```

#### Ici on peut avoir une estimation du Kmeans sur notre variable ce pendant les axes de notre pca ne capte que 13% de l'information de nos donnes l'ACP ne peut pas capter les relations non linaire, la reprsentation de cette dernire est donc fauss il nous faut faire des analyses sur d'autre espace qui capterons mieux les proprits de notre espace initial et qui nous donnerons si possible une meilleur rduction dimensionnelle 


```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from sklearn.cluster import KMeans

kmeans = KMeans()
kmeans.fit(X_new.A)
y_kmeans = kmeans.predict(X_new.A)
plt.scatter(X2D[:, 0], X2D[:, 1], c=y_kmeans)
plt.xlabel('composante 1')
plt.ylabel('composante 2')
plt.show();


```

#### estimation des diffrentes classe par l'algo EM

```python
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components=3)
gmm.fit(X_new.A)
y_gmm = gmm.predict(X_new.A)
plt.scatter(X2D[:, 0], X2D[:, 1], c=y_gmm)
plt.xlabel('composante 1')
plt.ylabel('composante 2')
plt.show();
```

#### PCA  noyaux

```python
from sklearn.decomposition import KernelPCA 
rdf_pca = KernelPCA(n_components = 2, kernel='rbf', gamma=0.04)
X2D_noy = rdf_pca.fit_transform(X_new.A)
kmeans = KMeans()
kmeans.fit(X_new.A)
y_kmeans = kmeans.predict(X_new.A)
plt.scatter(X2D_noy[:, 0], X2D_noy[:, 1], c=y_kmeans)
plt.xlabel('composante 1')
plt.ylabel('composante 2')
plt.show();
```
